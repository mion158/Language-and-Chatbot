```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = 'https://myanimelist.net/reviews.php?t=anime'
headers = {"User-Agent": "mal review scraper."}
response = get(url)
print(response.status_code)
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')
review_containers = html_soup.find_all('div', class_ = 'review-element js-review-element')
len(review_containers)
```

```python
#second review should be about One Piece
review = review_containers[1]
review
```

```python
recommend = review.find_all('div', attrs ={'class': 'tags'})[0].contents[0]
recommend
```

```python
recommend = review_containers[0].find_all('div', attrs ={'class': 'tag recommended btn-label js-btn-label'})[0].contents[1]
recommend
```

```python
#get anime name
names = []
for i in range(50):
  anime_name = review_containers[i].find_all('a', attrs ={'data-ga-click-type': 'review-anime-title'})[0].contents[0]
  names.append(str(anime_name))

print(names)
```

```python
#get anime rating
ratings = []
for i in range(50):
  anime_rating = review_containers[i].find_all('span', attrs ={'class': 'num'})[0].contents[0]
  ratings.append(int(anime_rating))

print(ratings)
```

```python
#get review content
contents = []
for i in range(50):
  anime_content = review_containers[i].find_all('div', attrs ={'class': 'text'})[0].contents[2:]
  anime_content = str(anime_content).replace('\\','').replace('<br/>','').replace('\'n', ' ').replace('\'r', ' ').replace('\"n', ' ')
  contents.append(anime_content)

print(contents)
```

```python
#get recommended?
recommended = []
for i in range(50):
  anime_recommend = review_containers[i].find_all('div', attrs ={'class': 'tags'})[0].contents[0]
  anime_recommend = str(anime_recommend)
  if ("</i>Recommended</div>" in anime_recommend):
    recommended.append(1)
  else:
    recommended.append(0)

print(recommended)
print(len(recommended))
```

```python
data = {"name": names, "rating": ratings, "content": contents, "recommended": recommended}
df = pd.DataFrame.from_dict(data)
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['content'], 
                                                    df['recommended'], 
                                                    random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Fit the CountVectorizer to the training data
vect = CountVectorizer().fit(X_train)
```

```python
vect.get_feature_names_out()[::2000]
```

```python
len(vect.get_feature_names_out())
```

```python
# transform the documents in the training data to a document-term matrix
X_train_vectorized = vect.transform(X_train)

X_train_vectorized
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_vectorized, y_train)
```

```python
from sklearn.metrics import roc_auc_score

predictions = model.predict(vect.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
# get the feature names as numpy array
feature_names = np.array(vect.get_feature_names_out())

# Sort the coefficients from the model
sorted_coef_index = model.coef_[0].argsort()

# Find the 10 smallest and 10 largest coefficients
print('weakest words:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
# indexed using [:-11:-1] so the list returned is in order of largest to smallest
print('strongest words: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]]))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Fit the TfidfVectorizer to the training data specifiying a minimum document frequency of 5
vect = TfidfVectorizer(min_df=5).fit(X_train)
len(vect.get_feature_names_out())
```

```python
X_train_vectorized = vect.transform(X_train)

model = LogisticRegression()
model.fit(X_train_vectorized, y_train)

predictions = model.predict(vect.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
feature_names = np.array(vect.get_feature_names_out())

sorted_tfidf_index = X_train_vectorized.max(0).toarray()[0].argsort()

print('weakest tfidf:\n{}\n'.format(feature_names[sorted_tfidf_index[:10]]))
print('strongest tfidf: \n{}'.format(feature_names[sorted_tfidf_index[:-11:-1]]))
```

```python
sorted_coef_index = model.coef_[0].argsort()

print('weakest Coefs:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
print('strongest Coefs: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]]))
```

```python
# Fit the CountVectorizer to the training data specifiying a minimum  document frequency of 5 and extracting 1-grams and 2-grams
vect = CountVectorizer(min_df=5, ngram_range=(1,2)).fit(X_train)

X_train_vectorized = vect.transform(X_train)

len(vect.get_feature_names_out())
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vectorized, y_train)

predictions = model.predict(vect.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
feature_names = np.array(vect.get_feature_names_out())

sorted_coef_index = model.coef_[0].argsort()

print('Smallest Coefs:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
print('Largest Coefs: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]]))
```
